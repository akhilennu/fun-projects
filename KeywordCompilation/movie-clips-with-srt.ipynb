```python
from moviepy.editor import *
import pysrt

IMAGEMAGICK_BINARY = os.getenv ('IMAGEMAGICK_BINARY', 'C:\Program Files\ImageMagick-7.1.0-Q16-HDRI\convert.exe')
```

```python
def extract_all_videos(video_path, parts):
    all_clips = []
    for part in parts:
        start, end = extract_start_end(part)
        all_clips.append(extract_video(video_path, start, end))
    return all_clips
def merge_all_clips(clips):
    return concatenate_videoclips(clips)
def extract_start_end(i):
    return subs[i].start.hours*3600+subs[i].start.minutes*60+subs[i].start.seconds, subs[i].end.hours*3600+subs[i].end.minutes*60+subs[i].end.seconds
def find_subtitle(subtitle, search_text, lo=0, size_diff=100, max_parts=10):
    i = lo
    ret = []
    while (i < len(subtitle)):
        if (subtitle[i].text.lower().find(search_text)!=-1 and len(subtitle[i].text) - len(search_text) <= size_diff and len(ret)<max_parts):
            ret.append(i)
        i+=1
    return ret
def extract_video(path, start, end):
    return VideoFileClip(path).subclip(start,end)
def display_text(subs, parts):
    for index, part in enumerate(parts):
        print(index, '|', part, "|", subs[part].text)
```

```python
base_path = "/media/akhilennu/Everything/movies/english/Animated/KungFuPanda"
srt_path = base_path+'/Kung.Fu.Panda.2008.720p.BluRay.x264. YTS.MX-English.srt'
video_path = base_path+"/Kung.Fu.Panda.2008.1080p.BluRay.x264-[YTS.AG].mp4"
search_word = "I don't know"
search_word = search_word.lower()
write_path = base_path+"/"+search_word+".mp4"
subs = pysrt.open(srt_path, encoding='iso-8859-1')
print("subtitles extracted")
parts = find_subtitle(subs,search_word, size_diff=15, max_parts=15)
print("found ", len(parts), "clips containing the keyword", search_word)
display_text(subs, parts)
```
